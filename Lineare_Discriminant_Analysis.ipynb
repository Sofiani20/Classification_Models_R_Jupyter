```python
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
df=pd.read_csv(r"C:\Users\dell\Documents\apprentissage_ML_DL_Python_R_udemy\codes\House_price_prepros_class.csv",header=0)
df
```

```python
X=df.loc[:,df.columns!='Sold']
X
```

```python
y=df['Sold']
y
```

```python
clf_lda=LinearDiscriminantAnalysis()
```

```python
clf_lda.fit(X,y)
```

```python
y_lda_pred=clf_lda.predict(X)
y_lda_pred
```

```python
confusion_matrix(y,y_lda_pred)
```

# TRAIN TEST


```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.2, random_state=0)
```

```python
clf_LDA=LinearDiscriminantAnalysis()
```

```python
clf_LDA.fit(X_train,y_train)
```

```python
y_test_pred=clf_LDA.predict(X_test)
```

```python
confusion_matrix(y_test,y_test_pred)
```

```python
accuracy_score(y_test,y_test_pred)
```
